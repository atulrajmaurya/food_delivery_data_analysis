```python
import pandas as pd
```

```python
df = pd.read_csv("final_food_delivery_dataset.csv")

df.head()
```

```python
gold_orders = df[df["membership"] == "Gold"].shape[0]
gold_orders
```

```python
hyd_revenue = round(df[df["city"] == "Hyderabad"]["total_amount"].sum())
hyd_revenue
```

```python
distinct_users = df["user_id"].nunique()
distinct_users
```

```python
avg_gold_order = round(
    df[df["membership"] == "Gold"]["total_amount"].mean(), 2
)
avg_gold_order
```

```python
high_rating_orders = df[df["rating"] >= 4.5].shape[0]
high_rating_orders
```

```python
gold_city_revenue = (
    df[df["membership"] == "Gold"]
    .groupby("city")["total_amount"]
    .sum()
)

top_gold_city = gold_city_revenue.idxmax()
top_gold_city
```

```python
gold_orders_top_city = df[
    (df["membership"] == "Gold") & (df["city"] == top_gold_city)
].shape[0]

gold_orders_top_city
```

```python
combo_revenue = (
    df.groupby(["membership", "cuisine"])["total_amount"]
    .sum()
    .sort_values(ascending=False)
)

combo_revenue
```

```python
df["order_date"] = pd.to_datetime(df["order_date"], dayfirst=True)
df["quarter"] = df["order_date"].dt.to_period("Q")

quarter_revenue = df.groupby("quarter")["total_amount"].sum()
quarter_revenue
```
